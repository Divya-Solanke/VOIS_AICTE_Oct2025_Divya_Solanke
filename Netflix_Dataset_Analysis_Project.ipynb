```python
df.drop_duplicates(inplace=True)
df.drop('Category',axis=1,inplace=True)
```

```python
print(df.isnull().sum())
```

```python
# Count of Movies vs TV Shows
sns.countplot(data=df, x='Category')
plt.title('Count of Movies and TV Shows')
plt.show()

# Top 10 countries with most content
top_countries = df['Country'].value_counts().head(10)
top_countries.plot(kind='bar', color='skyblue')
plt.title('Top 10 Content-Producing Countries')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.show()
```

```python
# Release_Date to datetime
df['Release_Date'] = pd.to_datetime(df['Release_Date'], errors='coerce')

# Year
df['Year'] = df['Release_Date'].dt.year

# Ploting count by year
sns.countplot(data=df, x='Year', order=sorted(df['Year'].dropna().unique()))
plt.title('Content Added Over the Years')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(data=df, x='Rating', order=df['Rating'].value_counts().index)
plt.title('Distribution of Ratings on Netflix')
plt.xticks(rotation=90)
plt.show()
```

```python
  # Filter only movies and fetch numeric duration
movies = df[df['Category'] == 'Movie']
movies['Minutes'] = movies['Duration'].str.replace(' min', '').astype(float)

sns.histplot(data=movies, x='Minutes', bins=30, kde=True)
plt.title('Movie Duration Distribution')
plt.xlabel('Minutes')
plt.show()
```

```python
# Split and count types
from collections import Counter
from itertools import chain

type_list = df['Type'].dropna().str.split(', ')
all_types = list(chain.from_iterable(type_list))
type_count = Counter(all_types).most_common(10)

# Convert to DataFrame for plotting
type_df = pd.DataFrame(type_count, columns=['Genre', 'Count'])

sns.barplot(data=type_df, x='Genre', y='Count', palette='mako')
plt.title('Top 10 Netflix Genres')
plt.xticks(rotation=45)
plt.show()
```

```python
top_countries = df['Country'].value_counts().head(5).index
subset = df[df['Country'].isin(top_countries)]

sns.countplot(data=subset, x='Country', hue='Category')
plt.title('Movies vs TV Shows by Country')
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
